```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from IPython.display import display, HTML
def show_info(data):
    tab_info=pd.DataFrame(data.dtypes).T.rename(index={0:'column type'})
    tab_info=tab_info.append(pd.DataFrame(data.isnull().sum()).T.rename(index={0:'null values (nb)'}))
    tab_info=tab_info.append(pd.DataFrame(data.isnull().sum()/data.shape[0]*100).T.rename(index={0:'null values(%)'}))
    return(display(tab_info))
```

```python
x = train.drop('class', axis =1)
y = train['class']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x,y, test_size = 0.3)
```

###  Modeling

WIthout any normalization/binning

```python
show_info(test)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import log_loss
from sklearn.metrics import accuracy_score
etc = ExtraTreesClassifier()
etc.fit(x_train,y_train)
y_pred = etc.predict(x_val)
print("Loss: ", log_loss(y_pred, y_val))
print("Accuracy: ", accuracy_score(y_pred, y_val))
```

```python
sub = pd.DataFrame(etc.predict_proba(test))
```
